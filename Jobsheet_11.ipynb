#Praktikum 01: Sequential Search pada Data Angka

```python
def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan di indeks {i}"
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

#Praktikum 02: Binary Search pada Data Angka

```python
def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Data {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

#Praktikum 03: Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
  for i in range(len(data)):
    if data[i].lower() == target.lower():
      return f"Nama {target} ditemukan di indeks {i}"
  return f"Nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

Praktikum 05: Perbandingan Waktu Sequential vs Binary Search

```python
import time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return i
  return -1

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print(f"Sequential Search waktu: {(end - start):<.8f} detik")

start = time.time()
binary_search(data, target)
end = time.time()
print(f"Binary Search waktu    : {(end - start):<.8f} detik")
```

#Praktikum 06: Kombinasi Sorting dan Searching

```python
import random

def insertion_sort(data):
  for i in range(1, len(data)):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"{target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = urut[7] # ambil angka ke-8 sebagai target
print(binary_search(urut, target))
```

#Penugasan 01: Variasi Sequential Search

```python
def sequential_search(data, target):
    perbandingan = 0

    for i in range(len(data)):
        perbandingan += 1  # setiap pengecekan dihitung
        if data[i] == target:
            return (
                f"Data {target} ditemukan di indeks {i}\n"
                f"Jumlah perbandingan: {perbandingan}"
            )

    return (
        f"Data {target} tidak ditemukan\n"
        f"Jumlah perbandingan: {perbandingan}"
    )


angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

#Penugasan 02: Variasi Binary Search

```python
def binary_search(data, target):
    left, right = 0, len(data) - 1
    langkah = 0

    while left <= right:
        langkah += 1  # setiap iterasi dihitung
        mid = (left + right) // 2

        if data[mid] == target:
            return (
                f"Data {target} ditemukan di indeks {mid}\n"
                f"Jumlah langkah (iterasi): {langkah}"
            )
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return (
        f"Data {target} tidak ditemukan\n"
        f"Jumlah langkah (iterasi): {langkah}"
    )


angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

#Penugasan 04: Studi Kasus Barang Toko

```python
# Bubble Sort berdasarkan harga
def bubble_sort_harga(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j][1] > data[j + 1][1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data


# Binary Search berdasarkan harga
def binary_search_harga(data, target_harga):
    left = 0
    right = len(data) - 1

    while left <= right:
        mid = (left + right) // 2

        if data[mid][1] == target_harga:
            return mid
        elif data[mid][1] < target_harga:
            left = mid + 1
        else:
            right = mid - 1

    return -1


# ===== DATA BARANG =====
barang = [
    ("Buku", 15000),
    ("Pensil", 5000),
    ("Tas", 75000),
    ("Pulpen", 10000),
    ("Penghapus", 3000)
]

print("Data awal:")
print(barang)

# ===== SORTING =====
bubble_sort_harga(barang)
print("\nData setelah diurutkan berdasarkan harga:")
print(barang)

# ===== BINARY SEARCH =====
cari_harga = int(input("\nMasukkan harga barang yang dicari: "))
index = binary_search_harga(barang, cari_harga)

if index != -1:
    print(f"Barang ditemukan: {barang[index][0]} dengan harga Rp {barang[index][1]}")
else:
    print("Barang dengan harga tersebut tidak ditemukan.")
```

#Penugasan 05: Analisis Kompleksitas

```python
import time

# Sequential Search
def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1


# Binary Search
def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


def test_search(n):
    data = list(range(n))
    target = n - 1  # worst case

    # Sequential Search timing
    start = time.perf_counter()
    sequential_search(data, target)
    seq_time = time.perf_counter() - start

    # Binary Search timing
    start = time.perf_counter()
    binary_search(data, target)
    bin_time = time.perf_counter() - start

    return seq_time, bin_time


# ===== PENGUJIAN =====
ukuran_data = [100, 1000, 10000]

print("Ukuran | Sequential (detik) | Binary (detik)")
print("-" * 45)

for n in ukuran_data:
    seq, bin_ = test_search(n)
    print(f"{n:<7}| {seq:<.8f} | {bin_:<.8f}")
```
