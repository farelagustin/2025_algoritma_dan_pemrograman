#Praktikum 01: Bubble Sort

```python
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah bubble sort:", bubble_sort(angka))
```

#Praktikum 02: Selection Sort

```python
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range(i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

#Praktikum 03: Insertion Sort

```python
def insertion_sort(data):
  n = len(data)
  for i in range(1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

angka = [12, 11, 13, 5, 6]
print('Data sebelum di urutkan:', angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

#Praktikum 04: Studi Kasus-Mengurutkan Nilai Mahasiswa

```python
def bubble_sort_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][1] < data[j + 1][1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

#Praktikum 06: Sorting Data Barang (Nama & Harga)

```python
def bubble_sort_key(data, key_index = 0, reverse = False):
  n = len(data)
  for i in range (n):
    for j in range (0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
      else:
          if data[j][key_index] > data[j + 1][key_index]:
            data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

#Penugasan 01: Eksperimen Bubble Sort (Descending)

```python
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j] < data[j + 1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah bubble sort:", bubble_sort(angka))
```

#Penugasan 02: Eksperimen Selection Sort

```python
def selection_sort_elemen_terbesar(data):
  n = len(data)
  for i in range(n - 1, 0, -1):
    max_index = 0
    for j in range(1, i + 1):
      if data[j] > data[max_index]:
        max_index = j
    data[i], data[max_index] = data[max_index], data[i]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort (elemen terbesar di akhir):", selection_sort_elemen_terbesar(angka))
```

#Penugasan 03: Eksperimen Insertion Sort

```python
def insertion_sort_nama_mahasiswa(data):
  n = len(data)
  for i in range(1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j].lower() > key.lower():
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data


nama_mahasiswa = ["Budi", "Andi", "Dewi", "Cici", "Eko"]

print("Data sebelum diurutkan:", nama_mahasiswa)
print("Data setelah diurutkan alfabetis:", insertion_sort_nama_mahasiswa(nama_mahasiswa))
```

#Penugasan 05: Analisis Kompleksitas

```python
import random

# Bubble Sort dengan penghitung perbandingan
def bubble_sort_count(data):
    count = 0
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            count += 1  # perbandingan
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return count


# Selection Sort dengan penghitung perbandingan
def selection_sort_count(data):
    count = 0
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            count += 1  # perbandingan
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return count


# Insertion Sort dengan penghitung perbandingan
def insertion_sort_count(data):
    count = 0
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0:
            count += 1  # perbandingan
            if data[j] > key:
                data[j + 1] = data[j]
                j -= 1
            else:
                break
        data[j + 1] = key
    return count


# Pengujian untuk ukuran data tertentu
def test_sort(n):
    data = list(range(n, 0, -1))  # worst case (terbalik)
    print(f"\nUkuran Data: {n}")

    print("Bubble Sort     :", bubble_sort_count(data.copy()), "perbandingan")
    print("Selection Sort  :", selection_sort_count(data.copy()), "perbandingan")
    print("Insertion Sort  :", insertion_sort_count(data.copy()), "perbandingan")

for ukuran in [10, 50, 100]:
    test_sort(ukuran)
```

#Penugasan 06: Proyek Mini

```python
def tampilkan_barang(barang):
    print("\nDaftar Barang:")
    for b in barang:
        print(f"Nama  : {b['nama']}")
        print(f"Harga : Rp {b['harga']}")
        print("-" * 30)

def input_barang():
    barang = []
    n = int(input("Masukkan jumlah barang: "))
    for i in range(n):
        print(f"\nBarang ke-{i+1}")
        nama = input("Nama barang  : ")
        harga = int(input("Harga barang : "))
        barang.append({"nama": nama, "harga": harga})
    return barang

def sort_barang(barang, kriteria, ascending=True):
    for i in range(1, len(barang)):
        key = barang[i]
        j = i - 1
        while j >= 0 and (
            (barang[j][kriteria] > key[kriteria] and ascending) or
            (barang[j][kriteria] < key[kriteria] and not ascending)
        ):
            barang[j + 1] = barang[j]
            j -= 1
        barang[j + 1] = key
    return barang

# ===== PROGRAM UTAMA =====
barang = input_barang()

while True:
    print("\nMENU SORTING:")
    print("1. Urutkan berdasarkan Nama (Ascending)")
    print("2. Urutkan berdasarkan Nama (Descending)")
    print("3. Urutkan berdasarkan Harga (Ascending)")
    print("4. Urutkan berdasarkan Harga (Descending)")
    print("0. Keluar")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        sort_barang(barang, "nama", True)
        print("\nBarang diurutkan berdasarkan nama (A-Z)")
        tampilkan_barang(barang)
    elif pilihan == "2":
        sort_barang(barang, "nama", False)
        print("\nBarang diurutkan berdasarkan nama (Z-A)")
        tampilkan_barang(barang)
    elif pilihan == "3":
        sort_barang(barang, "harga", True)
        print("\nBarang diurutkan berdasarkan harga (Termurah)")
        tampilkan_barang(barang)
    elif pilihan == "4":
        sort_barang(barang, "harga", False)
        print("\nBarang diurutkan berdasarkan harga (Termahal)")
        tampilkan_barang(barang)
    elif pilihan == "0":
        print("Program selesai.")
        break
    else:
        print("Pilihan tidak ada di menu!")
```
