#Praktikum 04: Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Total: ", jumlahkan(1,2,3,4,5))
biodata(Nama="Budi", Umur=20, Prodi='TRK')
```

#Praktikum 07: Fungsi Decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Sesudah fungsi dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

#Penugasan 08: Fungsi Decorator

```python
def decorator_halo(fungsi):
  def wrapper():
    print("Memulai fungsi...")
    fungsi()
    print("Fungsi selesai dijalankan.")
  return wrapper

@decorator_halo
def halo():
  print("Halo, Mahasiswa!")

halo()
```

#Penugasan 09: Studi Kasus Nyata - Program Nilai

```python
def input_nilai():
    nilaimahasiswa = []
    banyak = int(input("Masukkan jumlah mahasiswa: "))
    for i in range(banyak):
        n = int(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
        nilaimahasiswa.append(n)
    return nilaimahasiswa

nilaimahasiswa = input_nilai()

print("Nilai mahasiswa: ", nilaimahasiswa)


def hitung_rata(nilaimahasiswa):
  return sum(nilaimahasiswa) / len(nilaimahasiswa)

rata = hitung_rata(nilaimahasiswa)

print("Nilai rata-rata: ", rata)

def status_kelulusan(rata):
  if rata >= 60:
    return "Mahasiswa Lulus"
  else:
    return "Mahasiswa Tidak Lulus"

status = status_kelulusan(rata)

print("Status kelulusan: ", status)

nilai_akhir = lambda tugas, ujian: (tugas*0.4) + (ujian*0.6)

tugas = int(input("Masukkan nilai tugas: "))
ujian = int(input("Masukkan nilai ujian: "))
nilaiakhir = nilai_akhir(tugas, ujian)

print("Nilai akhir (40% Tugas, 60% Ujian): ", nilaiakhir)
```
