#Penugasan 01: Eksperimen dengan Modul math

```python
import math

def luas_lingkaran(r):
  return math.pi * (r*r)

luas1 = luas_lingkaran(7)
luas2 = luas_lingkaran(10)

print(f"Luas lingkaran (jari-jari 7)= {luas1:.2f}")
print(f"Luas lingkaran (jari-jari 10)= {luas2:.2f}")
```

#Penugasan 03: Modul Buatan Sendiri

```python
%%writefile konversi2.py

def celcius_ke_fahreinheit(c):
  f = c * 1.8 + 32
  print(f"Konversi dari {c}°C ke Fahrenheit adalah: {f:.2f}°F")

def fahreinheit_ke_celcius(f):
  c = (f - 32 ) / 1.8
  print(f"Konversi dari {f}°F ke Celcius adalah: {c:.2f}°C")
```

```python
import konversi2

konversi2.celcius_ke_fahreinheit(39)
konversi2.fahreinheit_ke_celcius(99)
```

#Penugasan 05: Studi Kasus Nilai

```python
import random
import laporan1

print("=========1 KUIS SEDERHANA =========")

jumlah_soal = int(input("Masukkan jumlah soal yang anda inginkan: "))
nilai_benar = 0

for i in range(jumlah_soal):
    a = random.randint(1, 11)
    b = random.randint(1, 11)

    print(f"Soal ke-{i+1}: {a} + {b} = ?")
    jawaban = int(input("Jawaban anda: "))

    if jawaban == a + b:
        print("Jawaban anda BENAR")
        nilai_benar += 1
    else:
        print("Jawaban anda SALAH")

laporan1.ringkasan(nilai_benar, jumlah_soal)
```

```python
%%writefile laporan1.py

def ringkasan(nilai_benar, total_soal):
    print("\n=========== HASIL KUIS ===========")
    print(f"Total soal      : {total_soal}")
    print(f"Jawaban benar   : {nilai_benar}")
    print(f"Skor akhir      : {(nilai_benar / total_soal) * 100:.2f}%")
    print("==================================")
```
